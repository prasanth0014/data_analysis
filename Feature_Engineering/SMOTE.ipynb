```python
from sklearn.datasets import make_classification

X,y=make_classification(n_samples=1000,n_features=2,n_redundant=0,n_clusters_per_class=1,weights=[0.90,0.10],random_state=42)
```

```python
import pandas as pd
X=pd.DataFrame(X,columns=["df1","df2"])
y=pd.DataFrame(y,columns=["target"])
final_df=pd.concat([X,y],axis=1)
final_df
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df["df1"],final_df["df2"],c=final_df["target"])
```
